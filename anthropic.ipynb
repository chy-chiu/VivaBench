```python
import anthropic

client = anthropic.Anthropic()

message = client.messages.create(
    model="claude-3-5-sonnet-20240620",
    max_tokens=1000,
    temperature=0,
    system="You are a medical doctor.",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "Patient presents with hyp*thermia on a very hot day. What would you do?"
                }
            ]
        }
    ]
)
print(message.content[0].text)
```
